# Extract morphological features in batch

```python
from anaASCFileFuncs import anaResults
from morphClassifier import loadDataMatrix
from batchExtractMorphFeatures import processEachFile
```

```python
# Extract data in batch from ASC files and exist .mat files containing morphological features analyzed. Then save the results into pickle files (.pk) in the same directory
dataMorphPath = 'dataMorph/asc&mat'
processEachFile(dataMorphPath,savePath='./dataMorph/pkFiles')
```

```python
# load the data from the saved .pk files and return the feature matrix and cell type labels.
# pkFile_path = './dataMorph/test'   
pkFile_path = './dataMorph/pkFiles'
data_Feat, data_label, varNames, pkFileNames = loadDataMatrix(pkFile_path)
```

```python
# sort and save the morphological feature matrix into a new pickle file.
cellNames=['NaN']*len(pkFileNames)
for i,each in enumerate(pkFileNames):
    # print(i)
    # print(each)
    # print(cellNames[i])
    cellNames[i] = each.split('.')[0]
    
morphFeatMatrix=pd.DataFrame(data_Feat,index=cellNames,columns=varNames)
morphFeatMatrix.insert(0,'cellTypeID',data_label.astype(int))

morphFeatMatrix.to_csv('./dataMorph/morphFeatMatrix.csv')
```
