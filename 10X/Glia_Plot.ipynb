```python
%pylab inline
import scanpy as sc
import pandas as pd
import seaborn as sns
import numpy as np
import scanpy as sc
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
adata = sc.read('./data_h5/Insular_sorted_clusters.h5ad')
Oligo_adata = sc.read('./data_h5/Oligo.h5ad')
Micro_adata = sc.read('./data_h5/Micro.h5ad')
OPC_adata = sc.read('./data_h5/OPC.h5ad')
Astro_adata = sc.read('./data_h5/astro.h5ad')
EnVL_adata = sc.read('./data_h5/EnVL.h5ad')
```

```python
figsize(3,3)
sc.pl.umap(adata,color=['cellTypeID','SNAP25'],legend_loc='on data',frameon=False)
sc.pl.umap(Oligo_adata,color=['leiden','MOG'],legend_loc='on data',frameon=False)
sc.pl.umap(Micro_adata,color=['leiden','C1QB'],legend_loc='on data',frameon=False)
sc.pl.umap(OPC_adata,color=['leiden','C1QL1'],legend_loc='on data',frameon=False)
sc.pl.umap(Astro_adata,color=['leiden','AQP4'],legend_loc='on data',frameon=False)
sc.pl.umap(EnVL_adata,color=['leiden','FLT1','BGN'],legend_loc='on data',frameon=False)
```

```python
adata_Glias = adata[adata.obs['cellTypeID_2'].isin([54,55,56,57,58,59]),]
```

```python
adata_Glias
```

```python
figsize(5,5)
sc.pl.umap(adata_Glias,color=['cellTypeID'],legend_loc='on data',frameon=False)
```

```python
cellTypeID = adata.obs['cellTypeID'].astype('str')
cellTypeID[adata.obs['cellTypeID_2']==59]='8'
adata.obs['cellTypeID'] = cellTypeID.astype('category')
figsize(4,4)
sc.pl.umap(adata,color=['cellTypeID'],legend_loc='on data',frameon=False)
```

```python
cellTypeID_colors = adata.uns['cellTypeID_colors']
cellTypeID_colors[4]='#E6E6E6'
cellTypeID_colors[5]='#DCDCDC'
cellTypeID_colors[6]='#DCDCDC'
cellTypeID_colors[7]='#FF00FF'
cellTypeID_colors[8]='#0000FF'
adata.uns['cellTypeID_colors'] = cellTypeID_colors
```

```python
figsize(5,5)
sc.pl.umap(adata,color=['cellTypeID'],legend_loc='right margin',frameon=False, save='allGliaUmap.pdf')
```

```python
markers = ['AQP4','C1QA','C1QL1','MOG','FLT1','BGN']
figsize(4,4)
cellTypeID_2_order = ['Astrocytes','Microglia','OPC','Oligocytes','Endothalial cells','VLMCs']
sc.pl.stacked_violin(adata_Glias, markers, groupby='cellType_labels_2',swap_axes = True, dendrogram=False, cmap='Reds', categories_order=cellTypeID_2_order,save='allGliaMarkervln.pdf')
```
